Classificação: Bagging
O Randon Florest, Floresta Aleatória, é um tipo de algoritmo de aprendizagem de máquIna flexível e de fácil manuseio.
Ele produz excelentes resultados, na maioria das vezes, mesmo sem a ajuda dos ajustes dos hiperparâmetros.
É um dos algoritmos mais fáceis de ser usados deviso a sua simplicidade e pelo simples fato de ser utilizados em tarefas
de classificações e de regressões.

```
import numpy as np
from matplotlib import pyplot as plt

ys = 200 + np.random.randn(100)
x = [x for x in range(len(ys))]

plt.plot(x, ys, '-')
plt.fill_between(x, ys, 195, where=(ys > 195), facecolor='g', alpha=0.6)

plt.title("Sample Visualization")
plt.show()
```

forest é bagging (enssacamento)
